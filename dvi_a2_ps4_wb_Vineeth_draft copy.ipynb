```python
# Count the number of matches per season and sort by season
season_counts = df_matches['season'].value_counts().sort_index()
 
# Calculate total matches and convert counts to percentages
total_matches = season_counts.sum()
season_percentages = (season_counts / total_matches) * 100
 
# Create a horizontal bar chart
plt.figure(figsize=(12, 6))
ax = sns.barplot(y=season_counts.index, x=season_percentages.values, color='skyblue', orient='h')
 
# Remove x-axis and unnecessary borders
sns.despine(top=True, right=True, bottom=True, left=True)

ax.xaxis.set_visible(False) 
plt.xlabel(None)  
plt.ylabel("Season", fontsize=12)
plt.title("Percentage of Matches in Each Season", fontsize=14, fontweight='bold')

for index, value in enumerate(season_percentages.values):
    ax.text(value +0.2, index, f"{value:.1f}%", va='center', fontsize=10, fontweight='bold', color='black') 
 
# Show the plot
plt.tight_layout()
plt.show()
```

### **Summary of Findings**  
1. The number of matches played in each season varies, with some seasons having a significantly higher proportion of matches.
2. The highest proportion of matches occurred in 2013, followed by 2012 and 2011, while 2009 had the lowest proportion.
3. The higher percentage of matches during 2011-2013 could be because of more number of teams participating in IPL during these seasons.  

### **Reason for Choosing a Horizontal Bar Chart**  
1. A **horizontal bar chart** improves readability when category labels (seasons) are long or numerous.  
2. It effectively represents categorical data (seasons) against numerical values (proportion of matches).
3. It provides a clear visual hierarchy, making it easier to compare values at a glance.

### **Pre-attentive Attributes Used**  
1. **Length:** The bar length directly corresponds to the proportion of matches, allowing for immediate comparison.
2. **Position:** Bars are aligned along a common axis, making differences between seasons immediately clear.  

### **Gestalt Principles Used**  
1. **Proximity:** Bars close together indicate they belong to the same dataset (seasons).  
2. **Similarity:** The bars are uniformly styled with the same width and orientation, helping viewers recognize them as part of the same category.

<h4>Explore the proportion (percentage) of winning scores - win_by_runs, win_by_wickets for the entire time period from 2008 to 2019. Visualize using a single chart<br><br>
    Write the python code in the below cell to create appropriate visual to perform the above task.
<h4>Answer in markdown cells below the visual <br><br>

```python
# Count total matches per season
total_matches_per_season = df_matches.groupby('season').size()

# Count number of matches won by runs and by wickets and convert to percentage of total matches per season
win_by_runs_count = df_matches.groupby('season')['win_by_runs'].apply(lambda x: (x > 0).sum())
win_by_wickets_count = df_matches.groupby('season')['win_by_wickets'].apply(lambda x: (x > 0).sum())

win_data_percentage = pd.DataFrame({
    'win_by_runs': (win_by_runs_count / total_matches_per_season) * 100,
    'win_by_wickets': (win_by_wickets_count / total_matches_per_season) * 100
})

# Create the line chart
plt.figure(figsize=(12, 6))
sns.set_style("white")
plt.plot(win_data_percentage.index, win_data_percentage['win_by_runs'], marker='o', linestyle='-', color='skyblue', alpha=0.8)
plt.plot(win_data_percentage.index, win_data_percentage['win_by_wickets'], marker='o', linestyle='-', color='lightcoral', alpha=0.8)

# Labels and title
plt.xlabel("Season", fontsize=12)
plt.ylabel("Percentage of Matches Won", fontsize=12)
plt.title("Trend of Wins by Runs vs. Wins by Wickets (2008-2019)", fontsize=14, fontweight='bold')
plt.xticks(range(2008, 2018), [str(year) for year in range(2008, 2018)], rotation=0)

# Ensure axis lines remain visible and grid lines are removed
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['left'].set_visible(True)
plt.gca().spines['bottom'].set_visible(True)

# Add text labels directly at the last point of each line
plt.text(win_data_percentage.index[-1], win_data_percentage['win_by_runs'].iloc[-1], 'Win by Runs', 
         color='skyblue', fontsize=12, verticalalignment='bottom', fontweight='bold')

plt.text(win_data_percentage.index[-1], win_data_percentage['win_by_wickets'].iloc[-1], 'Win by Wickets', 
         color='lightcoral', fontsize=12, verticalalignment='bottom', fontweight='bold')

# Show the plot
plt.tight_layout()
plt.show()
```

### **1. Summary of Findings**
1. On Analysing the dataset, we see that we have seasons data only between 2008 and 2017 and not until 2019.
2. The proportion of wins by wickets (red line) is generally higher in most seasons.
3. 2010 and 2015 had a higher percentage of wins by runs (light blue), making them exceptions.
4. 2013 stands out as the only season where both win types were exactly equal, showing a balance.
5. 2014 and 2016 had significantly more wins by wickets, with the highest peaks observed.
6. The trend fluctuates across seasons, with no long-term dominance of either win type.

### **2. Reason for Selecting a Line Chart**
1. A line chart is effective for visualizing trends over time (2008-2019).
2. It clearly shows fluctuations and comparisons between the two winning methods across multiple seasons.
3. Helps in identifying patterns, peaks, and dips across seasons, help identify where trends shift (e.g., 2010, 2013, 2014, 2015, and 2016).
4. The use of markers enhances clarity by highlighting exact data points.

### **3. Pre-Attentive Attributes Used**
1. **Color** - Different colors are used to distinguish between "Win by Runs" and "Win by Wickets."
2. **Position** - The data points and lines clearly show the rise and fall of the values over seasons.
3. **Line & markers** - The markers improve readability by making each data point distinct.

### **4. Gestalt Principles Used**
1. **Proximity** - The points connected by lines are grouped together, making it easy to follow trends.
2. **Continuity** - The smooth connection of points allows the viewer to see a continuous pattern, enhancing readability.
3. **Similarity** - The two lines share the same style (line thickness and markers), making it easy to compare trends despite different colors.

#### **Question** :
**How does the toss-winning team's decision (bat first or field first) influence their winning percentage at different venues? We are considering the top 10 Venues based on number of matches hosted.**


```python
# Select top 10 venues with the most matches
top_venues = df_matches['venue'].value_counts().head(10).index

# Filter data for only these venues
filtered_df = df_matches[df_matches['venue'].isin(top_venues)]

# Compute win percentage based on toss decision
toss_stats = (
    filtered_df.groupby(['venue', 'toss_decision'])['winner']
    .count()
    .unstack()
    .fillna(0)
)

toss_stats = toss_stats.div(toss_stats.sum(axis=1), axis=0) * 100

# Sort by 'Bat First' win percentage in descending order
toss_stats = toss_stats.sort_values(by="bat", ascending=False)

# Plot stacked bar chart
sns.set_style("white")  
fig, ax = plt.subplots(figsize=(12, 7))
toss_stats.plot(kind='barh', stacked=True, color=["lightcoral", "skyblue"], ax=ax)

# Add percentage labels inside the bars
for i, (bat, bowl) in enumerate(zip(toss_stats.iloc[:, 0], toss_stats.iloc[:, 1])):
    ax.text(bat / 2, i, f"{bat:.1f}", ha="center", va="center", fontsize=12, color="black")
    ax.text(bat + bowl / 2, i, f"{bowl:.1f}", ha="center", va="center", fontsize=12, color="black")

# Labels and title
plt.xlabel("Win Percentage (%)", fontsize=12)
plt.ylabel("Venue", fontsize=12)
plt.title("Win % of Toss-Winning Team by Venue & Decision", fontsize=14, fontweight="bold")
plt.tick_params(axis='x', which='both', bottom=False, labelbottom=False)

# Adjust legend position
plt.legend(["Bat First", "Bowl First"], title="Toss Decision", fontsize=12, title_fontsize=13, frameon=False, loc='upper right', bbox_to_anchor=(1.2, 1))

# Remove axis borders
sns.despine(left=True, bottom=True)
ax.grid(False)

# Show plot
plt.show()
```



### **1. Summarize your findings from the visual**

2. In most venues, **bowling first** tends to give a better win percentage.
